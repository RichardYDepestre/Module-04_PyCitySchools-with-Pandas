```python
# 4.7.5
# calculate average

def avg_reading_score():
    return school_data_complete_df["reading_score"].mean()

def avg_math_score():
    return school_data_complete_df["math_score"].mean()
    
# 4.7.6
# passing grade
def passing_math():
    return school_data_complete_df[ school_data_complete_df["math_score"] >= 70]

def passing_reading():
    return school_data_complete_df[school_data_complete_df["reading_score"] >= 70]

def passing_math_count():
    temp_df = passing_math()
    return temp_df["student_name"].count()

def passing_reading_count():
    temp_df = passing_reading()
    return temp_df["student_name"].count()

def passing_math_percentage():
    return passing_math_count() / float(student_count) * 100

def passing_reading_percentage():
    return passing_reading_count() / float(student_count) * 100

def passing_math_reading():
    df = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & 
                                 (school_data_complete_df["reading_score"] >= 70)]
    return df

def overall_passing_math_reading_count():
    df = passing_math_reading()
    o_count =  df["student_name"].count()
    return o_count

def overall_passing_percentage():
    o_count = overall_passing_math_reading_count()
    return o_count / student_count * 100

def create_summary_dataframe():
    district_summary_df = pd.DataFrame(
        [{"Total Schools": school_count,
          "Total Students": student_count,
          "Total Budget": schools_budget,
          "Average Math Score": avg_math_score(),
          "Average Reading Score": avg_reading_score(),
          "% Passing Math": passing_math_percentage(),
         "% Passing Reading": passing_reading_percentage(),
        "% Overall Passing": overall_passing_percentage()}]
    )
    return district_summary_df

district_summary_df = create_summary_dataframe()
district_summary_df
```
